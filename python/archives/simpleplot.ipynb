```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os, glob
import matplotlib.animation as animation
import glob, re

import scienceplots
plt.style.use('default')
plt.style.use(['science', 'high-vis', 'grid'])
plt.rcParams['figure.figsize'] = [5, 3]
```

```python
file_admc = "/home/remi/DEV/OptiDopi/build/SimpleSPAD_58.68_BreakdownProbability.csv"
file_mcintyre = "/home/remi/DEV/OptiDopi/build/POISSON_NL/McIntyre_58.56000.csv"
```

```python
x, admcbrp = np.loadtxt(file_admc, delimiter=',', unpack=True, skiprows=1)
xa, mcintyrebrp = np.loadtxt(file_mcintyre, delimiter=',', unpack=True, skiprows=1, usecols=(0,1))
```

```python
fig, ax = plt.subplots()
ax.plot(x, admcbrp, ".-", label="ADMC", lw=2)
ax.plot(xa, mcintyrebrp, label="McIntyre", lw=3)
ax.set_xlabel("$x$ ($\mu$m)")
ax.set_ylabel("Breakdown probability")
ax.legend()

fig.tight_layout()
fig.savefig("breakdown_probability.pdf")
fig.savefig("breakdown_probability.png", dpi=300)
```
