```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import matplotlib.animation as animation
import glob, re, os

import scienceplots
plt.style.use('default')
plt.style.use(['science', 'high-vis', 'grid'])
```

```python
def rastring(x):
    return 10*len(x) + np.sum(x**2 - 10*np.cos(2*np.pi*x))
    
```

```python
# 1D plot
x = np.linspace(-10.12, 10.12, 1000)
y = [rastring(np.array([xi])) for xi in x]
plt.plot(x, y)
```

```python
# plot the function
x = np.linspace(-5.12, 5.12, 4000)
y = np.linspace(-5.12, 5.12, 4000)
X, Y = np.meshgrid(x, y)
Z = rastring([X, Y])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_xlim(-5.12, 5.12)
ax.set_ylim(-5.12, 5.12)
ax.set_aspect('equal')
ax.imshow(Z, extent=[-5.12, 5.12, -5.12, 5.12], origin='lower', cmap='viridis')
ax.contour(X, Y, Z, 20, cmap='viridis')
plt.show()
```

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$z$')
ax.set_xlim(-5.12, 5.12)
ax.set_ylim(-5.12, 5.12)
ax.set_zlim(0, 100)
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
plt.show()
```
